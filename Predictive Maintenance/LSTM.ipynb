```python
import numpy as np
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.utils import resample
from sklearn.preprocessing import StandardScaler
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import LSTM, Dense, Dropout, BatchNormalization
from tensorflow.keras.callbacks import EarlyStopping, ReduceLROnPlateau
from tensorflow.keras.optimizers import Adam
from tensorflow.keras.utils import to_categorical
import time
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.metrics import confusion_matrix, classification_report
import torch

# Check if GPU is available
device = torch.device('cuda' if torch.cuda.is_available() else 'cpu')
print(f'Using device: {device}')

# Load dataset
data = pd.read_csv("standardized_features_multiclass2.csv")

# Balance the dataset
classes = data['label'].unique()
balanced_data = pd.DataFrame()

for cls in classes:
    class_data = data[data['label'] == cls]
    if len(class_data) < 48:
        class_data = resample(class_data, replace=True, n_samples=48, random_state=42)
    balanced_data = pd.concat([balanced_data, class_data])

# Shuffle the balanced dataset
balanced_data = balanced_data.sample(frac=1, random_state=42).reset_index(drop=True)

# Split features and labels
X = balanced_data.drop(columns=['label'])
y = balanced_data['label']

# Binary classification: Healthy vs Non-Healthy
y_binary = np.where(y == 'healthy', 1, 0)  # 1 = Healthy, 0 = Non-Healthy

# Feature extraction function (standardization with clipping)
def feature_extraction(X):
    scaler = StandardScaler()
    X_scaled = scaler.fit_transform(X)
    X_clipped = np.clip(X_scaled, -5, 5)  # Clip to avoid extreme values
    return X_clipped

# Feature extraction for train data
X_scaled = feature_extraction(X)

# Reshape data for LSTM (3D input)
X_scaled = X_scaled.reshape(X_scaled.shape[0], X_scaled.shape[1], 1)

# Split data 80% train, 20% test for binary classification
X_train, X_test, y_train_binary, y_test_binary = train_test_split(X_scaled, y_binary, test_size=0.2, stratify=y_binary, random_state=42)

# LSTM Model for Binary Classification with increased complexity
def build_lstm_binary():
    model = Sequential([
        LSTM(256, activation='tanh', input_shape=(X_train.shape[1], 1), return_sequences=True, kernel_regularizer='l2'),
        BatchNormalization(),
        Dropout(0.4),
        LSTM(128, activation='tanh', kernel_regularizer='l2'),
        BatchNormalization(),
        Dense(128, activation='relu', kernel_regularizer='l2'),
        Dropout(0.4),
        Dense(1, activation='sigmoid')
    ])
    optimizer = Adam(learning_rate=0.0003, clipnorm=1.0)  # Added gradient clipping
    model.compile(optimizer=optimizer, loss='binary_crossentropy', metrics=['accuracy'])
    return model

# Early stopping callback for binary classification
early_stopping_binary = EarlyStopping(monitor='val_loss', patience=10, restore_best_weights=True)
reduce_lr_binary = ReduceLROnPlateau(monitor='val_loss', factor=0.3, patience=3, min_lr=1e-6)

# Build and train the LSTM for binary classification
lstm_binary_model = build_lstm_binary()
lstm_binary_model.summary()

# Train the LSTM for Binary Classification
history_binary = lstm_binary_model.fit(
    X_train, y_train_binary, validation_data=(X_test, y_test_binary),
    epochs=200, batch_size=32, callbacks=[early_stopping_binary, reduce_lr_binary]
)

# Save the binary classification model
model_filename_binary = "best_binary_classification_lstm_model.h5"
lstm_binary_model.save(model_filename_binary)

# Evaluate the LSTM model for binary classification
test_time_start = time.time()
y_pred_binary = (lstm_binary_model.predict(X_test) > 0.5).astype(int)
test_time_end = time.time()
test_time_binary = test_time_end - test_time_start
print(f"Test Time (Binary Classification): {test_time_binary:.4f} seconds")

# Confusion Matrix for the test set (binary classification)
cm_binary = confusion_matrix(y_test_binary, y_pred_binary)

# Plot confusion matrix as heatmap (binary classification)
plt.figure(figsize=(10, 7))
sns.heatmap(cm_binary, annot=True, fmt='d', cmap='Blues')
plt.title(f"Binary Classification Confusion Matrix")
plt.xlabel("Predicted")
plt.ylabel("True")
plt.show()

# Classification report for binary classification
binary_report = classification_report(y_test_binary, y_pred_binary, target_names=['Non-Healthy', 'Healthy'])
print("Binary Classification Classification Report:\n")
print(binary_report)

# Plot training loss and accuracy
plt.figure(figsize=(12, 6))
plt.plot(history_binary.history['loss'], label='Train Loss')
plt.plot(history_binary.history['val_loss'], label='Validation Loss')
plt.plot(history_binary.history['accuracy'], label='Train Accuracy')
plt.plot(history_binary.history['val_accuracy'], label='Validation Accuracy')
plt.title('Training and Validation Metrics (Binary Classification)')
plt.xlabel('Epochs')
plt.ylabel('Metrics')
plt.legend()
plt.grid(True)
plt.show()

# Print performance metrics for binary classification
final_train_loss = history_binary.history['loss'][-1]
final_train_acc = history_binary.history['accuracy'][-1]
final_val_loss = history_binary.history['val_loss'][-1]
final_val_acc = history_binary.history['val_accuracy'][-1]
print("Final Train Loss (Binary):", final_train_loss)
print("Final Train Accuracy (Binary):", final_train_acc)
print("Final Validation Loss (Binary):", final_val_loss)
print("Final Validation Accuracy (Binary):", final_val_acc)
print(f"Binary Test Accuracy: {np.mean(y_pred_binary == y_test_binary):.4f}")

# Multi-class Classification: Now focusing on the 8 non-healthy classes (excluding 'healthy')

# Create multi-class target for non-healthy classes (excluding 'healthy')
y_multi = y[y != 'healthy']  # Exclude 'healthy'
X_multi = X_scaled[y != 'healthy']  # Also exclude corresponding rows in X
y_encoded_multi, label_mapping_multi = pd.factorize(y_multi)

# Convert labels to one-hot encoding for multi-class classification
y_encoded_multi_one_hot = to_categorical(y_encoded_multi)

# Split data for multi-class classification (excluding 'healthy')
X_train_multi, X_test_multi, y_train_multi, y_test_multi = train_test_split(
    X_multi, y_encoded_multi_one_hot, test_size=0.2, stratify=y_encoded_multi, random_state=42
)

# LSTM Model for Multi-Class Classification with increased complexity
def build_lstm_multi():
    model = Sequential([
        LSTM(256, activation='tanh', input_shape=(X_train_multi.shape[1], 1), return_sequences=True, kernel_regularizer='l2'),
        BatchNormalization(),
        Dropout(0.4),
        LSTM(128, activation='tanh', kernel_regularizer='l2', return_sequences=True),
        BatchNormalization(),
        Dropout(0.4),
        LSTM(64, activation='tanh', kernel_regularizer='l2'),
        BatchNormalization(),
        Dense(128, activation='relu', kernel_regularizer='l2'),
        Dropout(0.4),
        Dense(len(label_mapping_multi), activation='softmax')  # Adjust output layer size
    ])
    optimizer = Adam(learning_rate=0.0003, clipnorm=1.0)  # Added gradient clipping
    model.compile(optimizer=optimizer, loss='categorical_crossentropy', metrics=['accuracy'])
    return model

# Early stopping callback for multi-class classification
early_stopping_multi = EarlyStopping(monitor='val_loss', patience=10, restore_best_weights=True)
reduce_lr_multi = ReduceLROnPlateau(monitor='val_loss', factor=0.3, patience=3, min_lr=1e-6)

# Build and train the LSTM for multi-class classification
lstm_multi_model = build_lstm_multi()
lstm_multi_model.summary()

# Train the LSTM for Multi-Class Classification
history_multi = lstm_multi_model.fit(
    X_train_multi, y_train_multi, validation_data=(X_test_multi, y_test_multi),
    epochs=200, batch_size=32, callbacks=[early_stopping_multi, reduce_lr_multi]
)

# Save the multi-class classification model
model_filename_multi = "best_multi_classification_lstm_model.h5"
lstm_multi_model.save(model_filename_multi)

 #Evaluate the LSTM model for multi-class classification
test_time_start = time.time()
y_pred_multi = np.argmax(lstm_multi_model.predict(X_test_multi), axis=1)
test_time_end = time.time()
test_time_multi = test_time_end - test_time_start
print(f"Test Time (Multi-Class Classification): {test_time_multi:.4f} seconds")

# Confusion Matrix for the test set (multi-class classification)
cm_multi = confusion_matrix(np.argmax(y_test_multi, axis=1), y_pred_multi)

# Plot confusion matrix as heatmap (multi-class classification)
plt.figure(figsize=(10, 7))
sns.heatmap(cm_multi, annot=True, fmt='d', cmap='Blues')
plt.title(f"Multi-Class Classification Confusion Matrix")
plt.xlabel("Predicted")
plt.ylabel("True")
plt.show()

# Classification report for multi-class classification
print("Multi-Class Classification Classification Report:\n")
print(classification_report(np.argmax(y_test_multi, axis=1), y_pred_multi, target_names=label_mapping_multi))

# Print performance metrics for multi-class classification
accuracy_multi = np.mean(y_pred_multi == np.argmax(y_test_multi, axis=1))
print(f"Final Multi-Class Test Accuracy: {accuracy_multi:.4f} ({accuracy_multi*100:.2f}%)")

# Performance Metrics for Multi-Class Classification
print("\nMulti-Class Performance Metrics:")
print(f"Accuracy: {accuracy_multi:.4f} ({accuracy_multi*100:.2f}%)")
print(f"Test Time: {test_time_multi:.4f} seconds")

```

```python
import pandas as pd
import numpy as np
import time
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import LabelEncoder, StandardScaler
from sklearn.metrics import confusion_matrix, classification_report, accuracy_score
import tensorflow as tf
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import LSTM, Dense, Dropout
from tensorflow.keras.utils import to_categorical
from tensorflow.keras.callbacks import ModelCheckpoint, EarlyStopping
import seaborn as sns
import matplotlib.pyplot as plt

# Load dataset
file_path = '/content/standardized_features_multiclass2.csv'
data = pd.read_csv(file_path)

# Step 1: Binary Classification (Healthy vs Not Healthy)
data['binary_label'] = data['label'].apply(lambda x: 'healthy' if x == 'healthy' else 'not_healthy')
binary_label_encoded = LabelEncoder().fit_transform(data['binary_label'])
y_binary_one_hot = to_categorical(binary_label_encoded)

# Standardize features
scaler = StandardScaler()
X = scaler.fit_transform(data.drop(['label', 'binary_label'], axis=1).values)

X_train_binary, X_test_binary, y_train_binary, y_test_binary = train_test_split(
    X, y_binary_one_hot, test_size=0.2, stratify=binary_label_encoded, random_state=42
)

# Data Augmentation Function
def augment_data(X, y, augment_factor=3):
    augmented_X = []
    augmented_y = []
    for i in range(X.shape[0]):
        for _ in range(augment_factor):
            original = X[i]
            # Add random noise
            noise = np.random.normal(0, 0.02, original.shape)
            noisy_data = original + noise
            # Scale data
            scaled_data = original * np.random.uniform(0.9, 1.1, original.shape)
            # Collect augmented samples
            augmented_X.extend([original, noisy_data, scaled_data])
            augmented_y.extend([y[i]] * 3)
    return np.array(augmented_X), np.array(augmented_y)

# Apply Data Augmentation
X_train_binary_aug, y_train_binary_aug = augment_data(X_train_binary, y_train_binary, augment_factor=2)

# Reshape for LSTM input (3D shape: samples, time steps, features)
X_train_lstm_binary = X_train_binary_aug.reshape((X_train_binary_aug.shape[0], 1, X_train_binary_aug.shape[1]))
X_test_lstm_binary = X_test_binary.reshape((X_test_binary.shape[0], 1, X_test_binary.shape[1]))

# Learning Rate Scheduler
lr_schedule = tf.keras.optimizers.schedules.ExponentialDecay(
    initial_learning_rate=0.001,
    decay_steps=10000,
    decay_rate=0.9
)

# Define Binary LSTM Model
binary_lstm_model = Sequential([
    LSTM(128, input_shape=(X_train_lstm_binary.shape[1], X_train_lstm_binary.shape[2]), return_sequences=True),
    Dropout(0.3),
    LSTM(64, return_sequences=True),
    Dropout(0.3),
    LSTM(32, return_sequences=False),
    Dense(128, activation='relu'),
    Dropout(0.4),
    Dense(64, activation='relu'),
    Dropout(0.5),
    Dense(2, activation='softmax')  # Binary classification
])

binary_lstm_model.compile(optimizer=tf.keras.optimizers.Adam(learning_rate=lr_schedule),
                          loss='categorical_crossentropy', metrics=['accuracy'])

# Callbacks
binary_lstm_checkpoint = ModelCheckpoint('binary_lstm_model_optimized.keras', save_best_only=True, monitor='val_accuracy', mode='max')
early_stopping = EarlyStopping(monitor='val_loss', patience=15, restore_best_weights=True)

# Train the Model
binary_lstm_model.fit(
    X_train_lstm_binary, y_train_binary_aug, validation_split=0.2,
    epochs=100, batch_size=64, callbacks=[binary_lstm_checkpoint, early_stopping]
)

# Evaluate Model
binary_lstm_model = tf.keras.models.load_model('binary_lstm_model_optimized.keras')

start_time = time.time()
binary_preds = binary_lstm_model.predict(X_test_lstm_binary)
binary_classes = binary_preds.argmax(axis=1)
binary_accuracy = accuracy_score(y_test_binary.argmax(axis=1), binary_classes)
test_time = time.time() - start_time

# Print Final Metrics
print("--------------------------------------------------")
print(f"Final Test Accuracy: {binary_accuracy:.4f} ({binary_accuracy * 100:.2f}%)")
print(f"Test Time: {test_time:.4f} seconds")
print("\nBinary Classification Report:")
print(classification_report(y_test_binary.argmax(axis=1), binary_classes, target_names=['Healthy', 'Not Healthy']))

# Confusion Matrix
plt.figure(figsize=(8, 6))
sns.heatmap(confusion_matrix(y_test_binary.argmax(axis=1), binary_classes), annot=True, fmt='d', cmap='Blues',
            xticklabels=['Healthy', 'Not Healthy'], yticklabels=['Healthy', 'Not Healthy'])
plt.title('Binary Classification Confusion Matrix (Optimized)')
plt.xlabel('Predicted')
plt.ylabel('True')
plt.show()
```
